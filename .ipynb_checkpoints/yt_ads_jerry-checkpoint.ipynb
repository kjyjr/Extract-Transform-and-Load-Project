YT Ads Leaderboard: 2020 Year-End Wrap-Up

```python
yt_ads_json[0]
```

```python
yt_ads_2020 = yt_ads_json[0]
yt_ads_2020_df = pd.json_normalize(yt_ads_2020, 'data')
yt_ads_2020_df
```

```python
yt_ads_2020_df = yt_ads_2020_df.drop(['id', 'link_id', 'script_link'], axis=1)
yt_ads_2020_df
```

```python
reorg_yt_ads_2020_df = yt_ads_2020_df[['rank', 'brand', 'title', 'creative agency', 'media agency', 'href']]
reorg_yt_ads_2020_df
```

```python
reorg_yt_ads_2020_df['rank'] = reorg_yt_ads_2020_df['rank'] +1
top10_yt_ads_2020_df = reorg_yt_ads_2020_df
top10_yt_ads_2020_df
#https://stackoverflow.com/questions/70321252/how-do-i-increase-an-element-value-from-column-in-pandas
```

Observations on Top 10 YouTube ads for 2020:

- 3 were Super Bowl commercials
- 2 were for car makers (Hyundai, GMC)
- 2 (Nike and TurboTax) had the same Creative Agency (Wieden+Kennedy)
- 2 (Apple and NFL) had the same Media Agency (OMD)
- 1 (TurboTax) had the same Creative Agency and Media Agency (Wieden+Kennedy)
- 1 had the same Creative Agency as product maker (Apple)


```python
top10_yt_ads_2020 = top10_yt_ads_2020_df.to_dict('records')
pprint(top10_yt_ads_2020)
```

```python
mongo = MongoClient(port=27017)
```

```python
db = mongo.youtube_videos
top10_collection = db.top10_ads_2020
```

```python
top10_collection.insert_many(top10_yt_ads_2020)
```

```python
results = top10_collection.find()
for order in results:
    pprint(order)
```

Stay Home PSA Ads Leaderboard

```python
yt_ads_json[2]
```

```python
yt_psa_ads_2020 = yt_ads_json[2]
yt_psa_ads_2020_df = pd.json_normalize(yt_psa_ads_2020, 'data')
yt_psa_ads_2020_df
```

```python
yt_psa_ads_2020_df = yt_psa_ads_2020_df.drop(['id', 'link_id', 'script_link'], axis=1)
yt_psa_ads_2020_df
```

```python
reorg_yt_psa_ads_2020_df = yt_psa_ads_2020_df[['rank', 'brand', 'title', 'creative agency', 'media agency', 'href']]
reorg_yt_psa_ads_2020_df
```

```python
reorg_yt_psa_ads_2020_df['rank'] = reorg_yt_psa_ads_2020_df['rank'] +1
rank_yt_psa_ads_2020_df = reorg_yt_psa_ads_2020_df
rank_yt_psa_ads_2020_df
```

```python
top10_yt_psa_ads_2020_df = rank_yt_psa_ads_2020_df.iloc[0:10]
top10_yt_psa_ads_2020_df
```

Observations on Top 10 YouTube PSA ads for 2020:

- 2 of the brands in the Top 10 PSA ads for 2020 were also in the overall Top 10 ads for 2020 (NFL and Nike)
- 3 of the Top 10 PSA ads (Bud Light, Nike, and Old Spice) had the same Creative Agency (Wieden+Kennedy), an agency that also featured
  significantly in the overall Top 10 ads for 2020
- 2 of the Top 10 PSA ads (NFL, Pepsi) had the same Media Agency (OMD), an agency that also featured in 2 of the overall Top 10 ads
- For one PSA ad, Old Spice, both the Creative Agency and Media Agency were the same (Wieden+Kennedy)
- 4 of the PSA ads (Ad Council, Burger King, NFL, Oreo) emphasized the importance of staying home, and 2 (Ad Council, Pepsi), the theme of togetherness


```python
top10_yt_psa_ads_2020 = top10_yt_psa_ads_2020_df.to_dict('records')
pprint(top10_yt_psa_ads_2020)
```

```python
db = mongo.youtube_videos
top10_psa_collection = db.psa_top10_2020
```

```python
top10_psa_collection.insert_many(top10_yt_psa_ads_2020)
```

```python
results = top10_psa_collection.find()
for order in results:
    pprint(order)
```

YouTube Movie Trailers Leaderboard: Oscars 2020

```python
yt_ads_json[3]
```

```python
yt_trailers_2020 = yt_ads_json[3]
yt_trailers_2020_df = pd.json_normalize(yt_trailers_2020, 'data')
yt_trailers_2020_df
```

```python
yt_trailers_2020_df = yt_trailers_2020_df.drop(['id', 'link_id', 'script_link'], axis=1)
yt_trailers_2020_df
```

```python
reorg_yt_trailers_2020_df = yt_trailers_2020_df[['rank', 'title', 'brand', 'media agency', 'href']]
reorg_yt_trailers_2020_df
```

```python
reorg_yt_trailers_2020_df['rank'] = reorg_yt_trailers_2020_df['rank'] +1
top_yt_trailers_2020_df = reorg_yt_trailers_2020_df
top_yt_trailers_2020_df
```

Observations on Top YouTube Movie Trailers for 2020:

- Sony Pictures took 2 of the top trailers with Once Upon a Time in Hollywood and Little Women in spots 2 and 7, respectively. Both of those trailers
  also shared the same media agency, Universal McCann.
- Another media agency that figured significantly was OMD, taking the inner rankings of 4 and 6 for Jojo Rabbit and Ford v Ferrari, respectively. 
- Netflix also took 2 of the top trailers with The Irishman and Marriage Story ranked at 8 and 9, respectively.


```python
top_yt_trailers_2020 = top_yt_trailers_2020_df.to_dict('records')
pprint(top_yt_trailers_2020)
```

```python
db = mongo.youtube_videos
top_trailers_collection = db.top_trailers_2020
```

```python
top_trailers_collection.insert_many(top_yt_trailers_2020)
```

```python
results = top_trailers_collection.find()
for order in results:
    pprint(order)
```
